```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import scipy.stats as stats
from matplotlib.backends.backend_pdf import PdfPages

pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
path_train = '../data/train_set.csv'
data = pd.read_csv(path_train)
data.head()
data.columns
X = data.iloc[:,1:-2]
y = data.iloc[:,-2:]
```

# High resolution aggregate features

### hla_high_res_10 and granular features

```python
aggregate_feature = ['hla_high_res_10']
granular_features = ['hla_match_a_high', 'hla_match_b_high', 'hla_match_c_high', 
                     'hla_match_drb1_high', 'hla_match_dqb1_high']

df_selected = X[aggregate_feature + granular_features]

correlation_matrix = df_selected.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', 
            cbar_kws={'label': 'Correlation Coefficient'}, vmin=-1, vmax=1)

plt.title("Correlation Matrix: hla_high_res_10 vs Granular Features")

plt.show()
```

# Aggregate high resolution against low resolution features

```python

high_res_features = ['hla_high_res_10', 'hla_high_res_8', 'hla_high_res_6']
low_res_features = ['hla_low_res_10', 'hla_low_res_8', 'hla_low_res_6']

df_selected = X[high_res_features + low_res_features]

correlation_matrix = df_selected.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', 
            cbar_kws={'label': 'Spearman Correlation Coefficient'}, vmin=-1, vmax=1)

plt.title("Spearman Correlation Matrix: High-Res vs Low-Res Features")

plt.show()
```

# High resolution granular features against low resolution granular features

```python
high_res_granular_features = ['hla_match_a_high', 'hla_match_b_high', 'hla_match_c_high', 
                              'hla_match_drb1_high', 'hla_match_dqb1_high']

low_res_granular_features = [feature.replace('high', 'low') for feature in high_res_granular_features]

df_selected = X[high_res_granular_features + low_res_granular_features]

correlation_matrix = df_selected.corr(method='spearman')

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', 
            cbar_kws={'label': 'Spearman Correlation Coefficient'}, vmin=-1, vmax=1)

plt.title("Spearman Correlation Matrix: High-Res vs Low-Res Granular Features")

plt.show()
```
